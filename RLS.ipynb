# Regresión Lineal

```python
import pandas as pd #Manejo de datos y tabulación
import seaborn as sns #Mapas
import matplotlib.pyplot as plt #Mapas y base de sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Cargar los datos
data_path = 'Salary_Data.csv'
data = pd.read_csv(data_path)
```

```python
#Preparar los datos para una regresión
X = data[['YearsExperience']]
Y = data['Salary']
```

```python
# dividir los datos en sets de entrenamiento y de prueba
X_train,X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
#Crear el modelo de regresión lineal
model = LinearRegression()
```

```python
#Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)
```

```python
#Predecir los valores de salario para el conjunto de prueba
y_pred = model.predict(X_test)

#Calcular el Mean Squared Error (error cuadrático medio) y el coeficiente de determinación (R^2)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
#Gráfica de dispersión de los datos reales vs los predichos
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_test['YearsExperience'], y=y_test, color='blue', label='Actual')
sns.lineplot(x=X_test['YearsExperience'], y=y_pred, color='red', label='Predicted')
plt.title('Actual vs Predicted Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#Importar librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leer datos y ponerlos en un dataframe
df = pd.read_csv('Data.csv')
df
```

```python
#Crear gráfica
plt.figure(figsize=(10, 6)) #Dimensiones del gráfico
sns.scatterplot(data=df, x='Age', y='Salary', hue='Country', s=100, marker='^') #Datos a usar en Ejes X y Y, marcadores de país con forma triangular a escala de 100
plt.title('Distribución de Edad y Salario por País') #Titulo del gráfico
plt.xlabel('Edad') #Label del Eje X
plt.ylabel('Salario') #Label del Eje Y
plt.grid(True) #Grid de fondo
sns.set_theme(style='darkgrid', palette='bright') #Paleta de colores
plt.show() #Mostrar gráfico
```

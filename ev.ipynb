# Evaluación de Hipótesis

## Hipótesis

## Generación de Datos Simulados

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

plt.style.use('ggplot')

np.random.seed(0)

grupo_a = np.random.normal(100,10,100)
grupo_b = np.random.normal(110,10,100)
```

## Análisis Estadístico

```python
t_stat, p_valor = stats.ttest_ind(grupo_a, grupo_b)
alfa = 0.05

if p_valor < alfa:
  resultado = "Rechazamos la hipótesis nula (H0), hay diferencias significativas"
else:
  resultado = "No rechazamos la hipótesis nula (H0),no hay diferencias significativas"

print("Resultado del test t: " + resultado)
```

## Visualización de Resultados

```python
plt.figure(figsize=(10,6))
plt.hist(grupo_a, alpha=0.7, label='Grupo A')
plt.hist(grupo_b, alpha=0.7, label='Grupo B')
plt.title('Distribución de los Grupos A y B')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
#Importar librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
np.random.seed(0) #Semilla 0 para la generación de numeros aleatorios reproducibles
X = np.random.rand(100,2)
y = 3 + 2*X[:,0] + X[:,1] + np.random.randn(100) #La ultima parte es ruido

#Convertimos los datos a un DF para su manipulación
data = pd.DataFrame(data=X, columns=['Caracteristica_1','Caracteristica_2'])
data['Target'] = y
```

```python
data
```

```python
#Dividimos en sets de Entrenamiento y Prueba con un split 80/20
X_train, X_test, y_train, y_test = train_test_split(data[['Caracteristica_1','Caracteristica_2']], data['Target'], test_size=0.2, random_state=42)
```

```python
#Cargamos el Modelo de Regresión Lineal
model = LinearRegression()

#Entrenamos el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

#Realizar predicciones con el conjunto de prueba basandonos en el modelo entrenado
predictions = model.predict(X_test)

#Se calcula la precisión del modelo
score = model.score(X_test, y_test)
```

```python
print(str(score))
```

```python
#Visualizar la relación entre las caracteristicas y el objetivo
sns.pairplot(data, x_vars=['Caracteristica_1','Caracteristica_2'], y_vars='Target', kind='reg', height=5)
plt.show()
```

```python
#Graficar los resultados de la Regresión Lineal Múltiple
plt.figure(figsize=(10,5))

np.random.seed(42) #Semilla 42 para resultados reproducibles
y_test_sample = np.random.rand(20) * 10 #Valores reales simulados
predictions_sample = y_test_sample * 0.9 + np.random.rand(20) * 0.5 # Predicciones simuladas

plt.scatter(y_test_sample, predictions_sample, color='blue', label='Predictions vs Actual')
plt.plot([y_test_sample.min(), y_test_sample.max()], [y_test_sample.min(), y_test_sample.max()], 'k--', lw=2, color='red', label='Ideal Line')

plt.xlabel('Actual Values')
plt.ylabel('Predictions')
plt.title('Actual vs Predicted Values - Simulation')
plt.legend()
plt.show()
```

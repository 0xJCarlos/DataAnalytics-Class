# Análisis de Distribución de Probabilidad

## Generación de Datos Simulados

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar datos simulados siguiendo una distribución normal
np.random.seed(0)
datos = np.random.normal(0,1,1000) #1000 puntos con media 0 y desviación estandar 1
```

## Análisis de la Distribución

```python
#Gráfico del histograma de los datos
plt.figure(figsize=(10,6))
plt.hist(datos, bins=30, density=True, alpha=0.6, color='g')

#Superponemos la curva de densidad
from scipy.stats import norm
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit de la distribución normal"
plt.title(title)
plt.show()
```
